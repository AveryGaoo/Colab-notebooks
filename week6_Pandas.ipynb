```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# 1) 垂直拼接 (axis=0，默认) -> 行变多
result_rows = pd.concat([df1, df2], axis=0)
print(result_rows)
# 结果行是 1,2,5,6 这样往下接

print("")

# 2) 水平拼接 (axis=1) -> 列变多
result_cols = pd.concat([df1, df2], axis=1)
print(result_cols)
# 结果会有 A,B,A,B 两套列
```

🔹 pandas.concat() 是“智能拼接”，带索引自动对齐；

🔹 numpy.concatenate() 是“机械拼接”，纯按位置拼接。

```python
left = pd.DataFrame({
    'key': ['K0', 'K1', 'K2'],
    'A': ['A0', 'A1', 'A2']
})

right = pd.DataFrame({
    'key': ['K0', 'K1', 'K3'],
    'B': ['B0', 'B1', 'B2']
})

# 内连接 inner join：只要两边都出现过的key才留下
inner_result = pd.merge(left, right, on='key', how='inner')
print(inner_result)
print("")

# 外连接 outer join：所有key都要，不管有没有匹配，没的地方是NaN
outer_result = pd.merge(left, right, on='key', how='outer')
print(outer_result)
print("")

# 左连接 left join：保留left全部，能匹配right的就补上
left_result = pd.merge(left, right, on='key', how='left')
print(left_result)
print("")

# 右连接 right join：保留right全部
right_result = pd.merge(left, right, on='key', how='right')
print(right_result)
print("")
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2']},
                   index=['K0', 'K1', 'K2'])
df2 = pd.DataFrame({'B': ['B0', 'B1', 'B2']},
                   index=['K0', 'K2', 'K3'])

inner_j = df1.join(df2, how='inner')   # 只保留共同的 index
print(inner_j)
print("")

outer_j = df1.join(df2, how='outer')   # 保留所有 index
print(outer_j)
print("")
```

```python
import pandas as pd
import numpy as np
from tabulate import tabulate

# 构造一个示例 DataFrame，里面故意放了缺失值 NaN
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, None, 22, None, 30],          # 有缺失
    'City': ['Lisbon', 'Porto', None, 'Lisbon', 'Coimbra'],
    'Salary': [2000, 2500, None, 2200, None]  # 有缺失
})

print("原始数据：")
print(tabulate(df, headers='keys', tablefmt='psql'))
print('')

print("哪些位置是缺失值(True=缺失)：")
print(df.isnull())
print('')

print("\n每一列有多少缺失值：")
print(df.isnull().sum())
print('')
print('')
```

```python
df_fill_constant = df.copy()

#链式赋值 写法未来不再推荐: df_fill_constant['Age'].fillna(0, inplace=True)
df_fill_constant['Age'] = df_fill_constant['Age'].fillna(0)       #用0填Age的缺失
df_fill_constant['City'] = df_fill_constant['City'].fillna('UKK')
df_fill_constant['Salary'] = df_fill_constant['Salary'].fillna(0)

#如下：用 DataFrame 一次填多列，也可以：
#df_fill_constant.fillna({'Age': 0, 'Salary': 0}, inplace=True)

print("用固定值0填充缺失后：")
print(tabulate(df_fill_constant, headers='keys', tablefmt='psql'))
print('')
```

```python
df_ffill = df.copy()

# 用上一行的值填Age缺失， 场景：时间序列很常用，比如：今天价格缺失，用昨天的价格顶上
df_ffill['Age'] = df_ffill['Age'].ffill() #.ffill()前一行顶上(ffill是forward fill的简写)
                                          #.bfill()后一行顶上
# 用上一行的值填City缺失
df_ffill['City'] = df_ffill['City'].ffill()

# 用上一行的值填Salary缺失
df_ffill['Salary'] = df_ffill['Salary'].ffill()

print("使用前一行的值向下填充(ffill)后：")
print(tabulate(df_ffill, headers='keys', tablefmt='psql'))
# 注意：第0行没前一行，所以 0 行永远不会被 ffill 改写。
```

```python
#一次性写所有列 -> 对整个 DataFrame 都执行前向填充，可以直接这样写：
df_ffill_all = df.copy()
df_ffill_all = df_ffill_all.ffill()
print(tabulate(df_ffill_all, headers='keys', tablefmt='psql'))
```

```python
drop_row = df.copy()
drop_row2 = df.copy()
drop_row3 = df.copy()
drop_row4 = df.copy()

#只要这一行里有一个NaN，这行就删掉;
drop_row.dropna(inplace=True)
print(tabulate(drop_row, headers='keys', tablefmt='psql'))
print('')

#当整行全是NaN,才删;
drop_row2.dropna(how='all', inplace=True)
print(tabulate(drop_row2, headers='keys', tablefmt='psql'))
print('')

#有缺失就把整列干掉;
drop_row3.dropna(axis=1, inplace=True)
print(tabulate(drop_row3, headers='keys', tablefmt='psql'))
print('')

#整列全是NaN才删
drop_row4.dropna(axis=1, how='all', inplace=True)
print(tabulate(drop_row4, headers='keys', tablefmt='psql'))
print('')
```

```python
#数据类型转换 & 清洗格式: 现实世界的数据经常是“脏”的，
#比如€1,200.00这种字符串;你必须先清洗成数字，才能做计算画图跑模型;
trans_df = df.copy()
trans_df['Age'] = trans_df['Age'].ffill()
trans_df['Age'] = pd.to_numeric(trans_df['Age'], errors='coerce')   #errors='coerce'：转换不了的值会变成NaN，不会报错；


print(tabulate(trans_df.sort_values('Age'), headers='keys', tablefmt='psql'))  #sort_values('col',inplace=默认False)：原表不动，返回新表；
print("")                                                                      #sort_values('col',inplace=True）：直接在原表上排序,不返回新表(即：不再有输出)；

print(tabulate(trans_df, headers='keys', tablefmt='psql'))                     #可以看出，原表并没有被改动，只有copy的新表排序变动了
```

| 方法名                     | 用于               | 排序依据 | 示例                      |
| ----------------------- | ---------------- | ---- | ----------------------- |
| `list.sort()`           | Python 列表        | 元素值  | `[3,1,2].sort()`        |
| `sorted(list)`          | Python 列表（返回新对象） | 元素值  | `sorted([3,1,2])`       |
| `df.sort_values('col')` | pandas DataFrame | 列值   | `df.sort_values('Age')` |
| `df.sort_index()`       | pandas DataFrame | 索引号  | `df.sort_index()`       |
| `df.sort()`             | ❌ 废弃             | ——   | ❌ 报错（老语法）               |
