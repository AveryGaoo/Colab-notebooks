```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# 1) å‚ç›´æ‹¼æ¥ (axis=0ï¼Œé»˜è®¤) -> è¡Œå˜å¤š
result_rows = pd.concat([df1, df2], axis=0)
print(result_rows)
# ç»“æœè¡Œæ˜¯ 1,2,5,6 è¿™æ ·å¾€ä¸‹æ¥

print("")

# 2) æ°´å¹³æ‹¼æ¥ (axis=1) -> åˆ—å˜å¤š
result_cols = pd.concat([df1, df2], axis=1)
print(result_cols)
# ç»“æœä¼šæœ‰ A,B,A,B ä¸¤å¥—åˆ—
```

ğŸ”¹ pandas.concat() æ˜¯â€œæ™ºèƒ½æ‹¼æ¥â€ï¼Œå¸¦ç´¢å¼•è‡ªåŠ¨å¯¹é½ï¼›

ğŸ”¹ numpy.concatenate() æ˜¯â€œæœºæ¢°æ‹¼æ¥â€ï¼Œçº¯æŒ‰ä½ç½®æ‹¼æ¥ã€‚

```python
left = pd.DataFrame({
    'key': ['K0', 'K1', 'K2'],
    'A': ['A0', 'A1', 'A2']
})

right = pd.DataFrame({
    'key': ['K0', 'K1', 'K3'],
    'B': ['B0', 'B1', 'B2']
})

# å†…è¿æ¥ inner joinï¼šåªè¦ä¸¤è¾¹éƒ½å‡ºç°è¿‡çš„keyæ‰ç•™ä¸‹
inner_result = pd.merge(left, right, on='key', how='inner')
print(inner_result)
print("")

# å¤–è¿æ¥ outer joinï¼šæ‰€æœ‰keyéƒ½è¦ï¼Œä¸ç®¡æœ‰æ²¡æœ‰åŒ¹é…ï¼Œæ²¡çš„åœ°æ–¹æ˜¯NaN
outer_result = pd.merge(left, right, on='key', how='outer')
print(outer_result)
print("")

# å·¦è¿æ¥ left joinï¼šä¿ç•™leftå…¨éƒ¨ï¼Œèƒ½åŒ¹é…rightçš„å°±è¡¥ä¸Š
left_result = pd.merge(left, right, on='key', how='left')
print(left_result)
print("")

# å³è¿æ¥ right joinï¼šä¿ç•™rightå…¨éƒ¨
right_result = pd.merge(left, right, on='key', how='right')
print(right_result)
print("")
```

```python
import pandas as pd
import numpy as np
from tabulate import tabulate

# æ„é€ ä¸€ä¸ªç¤ºä¾‹ DataFrameï¼Œé‡Œé¢æ•…æ„æ”¾äº†ç¼ºå¤±å€¼ NaN
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, None, 22, None, 30],          # æœ‰ç¼ºå¤±
    'City': ['Lisbon', 'Porto', None, 'Lisbon', 'Coimbra'],
    'Salary': [2000, 2500, None, 2200, None]  # æœ‰ç¼ºå¤±
})

print("åŸå§‹æ•°æ®ï¼š")
print(tabulate(df, headers='keys', tablefmt='psql'))
print('')

print("å“ªäº›ä½ç½®æ˜¯ç¼ºå¤±å€¼(True=ç¼ºå¤±)ï¼š")
print(df.isnull())
print('')

print("\næ¯ä¸€åˆ—æœ‰å¤šå°‘ç¼ºå¤±å€¼ï¼š")
print(df.isnull().sum())
print('')
print('')
```
