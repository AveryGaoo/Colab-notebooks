```python
#In the Robot class, I use the constructor to initialize each robot with a name, its starting energy,
#and an efficiency value. I also set the initial distance to zero, because the robot hasn’t moved yet.
#In the move method, I calculate how much energy the robot needs based on the speed.
# The energy cost is speed squared divided by the robot’s efficiency.
# If the robot has enough energy, I subtract that cost and update the total distance the robot has travelled.
class Robot:
    def __init__(self, name, energy, efficiency):
        self.name = name
        self.energy = energy
        self.efficiency = efficiency
        self.distance = 0

    def move(self, speed):
        cost = speed**2 / self.efficiency
        if self.energy >= cost:
            self.energy -= cost
            self.distance += speed
```

```python
#PART 2 Robot Competition

"""
我定义了一个 RobotCompetition 类，在构造函数里接收多个机器人，并把它们保存起来。
Here I define a class called RobotCompetition.
In the constructor, I allow multiple robots to be passed in, and I store them in the robots attribute.

关于 start_race 方法：
在这个方法里，我通过让每个机器人以速度 10 移动多轮，来模拟比赛，每轮代表一次移动。
For the start_race method:
In the start_race method, I simulate the race by letting every robot move at a constant speed of 10 for a number of steps.
Each step represents one movement cycle.

关于决出胜者：
比赛结束后，我比较所有机器人的行驶距离。先假设第一个机器人是赢家，然后依次检查其他机器人，如果谁跑得更远，就更新赢家。最后打印出胜利者的名字和跑出的距离。
For determining the winner:
After all movements are finished, I compare the distances of all robots.
I start by assuming the first robot is the winner, and then I check each of the remaining robots. If any robot has travelled farther, I update the winner.
Finally, I print the name and distance of the robot that won the race.
"""
class RobotCompetition:
    def __init__(self, *robots):
        self.robots = robots

    def start_race(self, steps=10):
        for step in range(steps):
            for r in self.robots:
                r.move(10)  # constant speed

        # Determine the winner without lambda
        winner = self.robots[0]
        for r in self.robots[1:]:
            if r.distance > winner.distance:
                winner = r

        print(f"Winner: {winner.name} ({winner.distance:.1f}m)")

```

```python
#2.1. Create three robots
RobotA = Robot("Tó", 100, 8)
RobotB = Robot("Xico", 100, 1)
RobotC = Robot("Manel", 100, 5)
Robots = RobotCompetition(RobotA, RobotB, RobotC)
```

```python
#2.2. Run a 10-step race  2.3. Which robot won?
Robots.start_race()
print("")

RobotA.print_info()
RobotB.print_info()
RobotC.print_info()
```

```python
#PART 3 Robot with Randomness
import random

class RandomRobot(Robot):
    def move(self, speed):
        # Add a small adjustment to speed: ±10%
        adjustment = random.uniform(0.9, 1.1)
        adjusted_speed = speed * adjustment

        # Optional: also adjust efficiency slightly
        adjusted_efficiency = self.efficiency * random.uniform(0.95, 1.05)

        cost = adjusted_speed**2 / adjusted_efficiency
        if self.energy >= cost:
            self.energy -= cost
            self.distance += adjusted_speed

        return adjusted_speed
```
