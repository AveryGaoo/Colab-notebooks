```python
import pandas as pd
data = {
'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
'Age': [25, 30, 35, 40, 28],
'City': ['Lisbon', 'Porto', 'Faro', 'Coimbra', 'Faro'],
'Salary': [2500, 3000, 4000, 3500, 2800]
}
df = pd.DataFrame(data)
df
```

```python
df[df['Age'] > 30]
```

```python
df[df['Age'] > 35]
```

```python
df[(df['Age'] > 30) & (df['Salary'] > 3500)]
```

```python
#Task 1: Find employees who are younger than 35 and earn more than 2700.
df[(df['Age'] < 35) & (df['Salary'] > 2700)]
```

```python
df.query('Age > 30 and Salary > 3000')
```

```python
#Now try using a variable inside a query:
limit = 2800
df.query('Salary >= @limit')
```

```python
#Task 2: Use .query() to find employees younger than 35 who work in Faro.
limit = 'Faro'
df.query('Age < 35 and City == @limit')
# df.query("Age < 35 and City  == 'Faro'")
#df[(df['Age'] < 35) & (df['City'] == 'Faro')]
```

```python
#Task 3: Write a filter that shows employees NOT from Faro or Porto. 
#(Hint: use the ~ symbol before df['City'].isin(...))
df[~df['City'].isin(['Faro', 'Porto'])]
```

```python
df[df['Age'].between(30, 40)]
```

```python
#Task 4: Find employees whose salaries are between 2800 and 3500.
df[df['Salary'].between(2800, 3500)]
```

```python
df['FilteredSalary'] = df['Salary'].where(df['Salary'] > 3000)
df['MaskedSalary'] = df['Salary'].mask(df['Salary'] > 3500)
df
```
